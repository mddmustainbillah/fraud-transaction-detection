```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame named 'df'
# If not, you can load it using: df = pd.read_csv('your_data.csv')

# Create a histogram for the entire dataset
plt.figure(figsize=(12, 6))
plt.hist(df['distance_from_home'], bins=100, color='blue', edgecolor='black')
plt.title('Distribution of Distance from Home for All Transactions')
plt.xlabel('Distance from Home')
plt.ylabel('Number of Transactions')
plt.show()

# Separate histograms for fraud and non-fraud transactions
fraud_data = df[df['fraud'] == 1]
non_fraud_data = df[df['fraud'] == 0]

plt.figure(figsize=(12, 6))

# Plotting histogram for fraud transactions
plt.subplot(1, 2, 1)
plt.hist(fraud_data['distance_from_home'], bins=200, color='red', edgecolor='black')
plt.title('Distribution of Distance from Home for Fraud Transactions')
plt.xlabel('Distance from Home')
plt.ylabel('Number of Transactions')

# Plotting histogram for non-fraud transactions
plt.subplot(1, 2, 2)
plt.hist(non_fraud_data['distance_from_home'], bins=200, color='green', edgecolor='black')
plt.title('Distribution of Distance from Home for Non-Fraud Transactions')
plt.xlabel('Distance from Home')
plt.ylabel('Number of Transactions')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame named 'df'
# If not, you can load it using: df = pd.read_csv('your_data.csv')

# Step 1: Calculate average, min, and max distance_from_home for fraud and non-fraud transactions
fraud_data = df[df['fraud'] == 1]
non_fraud_data = df[df['fraud'] == 0]

fraud_stats = fraud_data['distance_from_home'].agg(['mean', 'min', 'max'])
non_fraud_stats = non_fraud_data['distance_from_home'].agg(['mean', 'min', 'max'])

# Step 2: Create separate histograms for fraud and non-fraud transactions
plt.figure(figsize=(12, 6))

# Plotting histogram for fraud transactions
plt.subplot(1, 2, 1)
plt.hist(fraud_data['distance_from_home'], bins=[0, 50, 100, 150, 200, 250], color='red', edgecolor='black')
plt.title('Fraud Transactions')
plt.xlabel('Distance from Home')
plt.ylabel('Number of Transactions')

# Plotting histogram for non-fraud transactions
plt.subplot(1, 2, 2)
plt.hist(non_fraud_data['distance_from_home'], bins=[0, 50, 100, 150, 200, 250], color='green', edgecolor='black')
plt.title('Non-Fraud Transactions')
plt.xlabel('Distance from Home')
plt.ylabel('Number of Transactions')

plt.tight_layout()
plt.show()
```
